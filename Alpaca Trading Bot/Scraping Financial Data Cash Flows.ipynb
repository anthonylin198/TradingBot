```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as ur
```

```python
# Enter a stock symbol
index= "FB"
# URL link 
url_is = "https://finance.yahoo.com/quote/" + index + "/financials?p=" + index
url_bs = "https://finance.yahoo.com/quote/" + index +"/balance-sheet?p=" + index
url_cf = "https://finance.yahoo.com/quote/" + index + "/cash-flow?p="+ index
```

```python
read_data = ur.urlopen(url_cf).read() 
soup_is= BeautifulSoup(read_data,"lxml")
```

```python
ls= [] # Create empty list
for l in soup_is.find_all("div"): 
  #Find all data structure that is ‘div’
  ls.append(l.string) # add each element one by one to the list
```

```python
new_ls = list(filter(None,ls))
```

```python
new_ls = new_ls[12:]
```

```python
is_data = list(zip(*[iter(new_ls)]*6))
```

```python
cashFlow = pd.DataFrame(is_data[0:])
```

```python
cashFlow.columns = cashFlow.iloc[0] # Name columns to first row of dataframe
cashFlow = cashFlow.iloc[1:,] # start to read 1st row
cashFlow = cashFlow.T # transpose dataframe
cashFlow.columns = cashFlow.iloc[0] #Name columns to first row of dataframe
cashFlow.drop(cashFlow.index[0],inplace=True) #Drop first index row
cashFlow.index.name = "" # Remove the index name
cashFlow.rename(index={"ttm": "12/31/2019"},inplace=True) #Rename ttm in index columns to end of the year
cashFlow = cashFlow[cashFlow.columns[:-5]] # remove last 5 irrelevant columns
```
