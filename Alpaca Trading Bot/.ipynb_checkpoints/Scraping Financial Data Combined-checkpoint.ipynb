```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as ur
```

```python
# Enter a stock symbol
index= "GRUB"
# URL link 
url_is = "https://finance.yahoo.com/quote/" + index + "/financials?p=" + index
url_bs = "https://finance.yahoo.com/quote/" + index +"/balance-sheet?p=" + index
url_cf = "https://finance.yahoo.com/quote/" + index + "/cash-flow?p="+ index
```

```python
# read the income statement
read_data = ur.urlopen(url_is).read() 
soup_is= BeautifulSoup(read_data,"lxml")

# read in the balance sheet
read_data_bs = ur.urlopen(url_bs).read() 
soup_bs= BeautifulSoup(read_data_bs,"lxml")

# read in the statement of cash flows
read_data_cf = ur.urlopen(url_cf).read() 
soup_cf= BeautifulSoup(read_data_cf,"lxml")
```

```python
# Creating the list for the income statement
ls= [] # Create empty list
for l in soup_is.find_all("div"): 
  #Find all data structure that is ‘div’
  ls.append(l.string) # add each element one by one to the list
 
ls = [e for e in ls if e not in ("Operating Expenses","Non-recurring Events")] # Exclude those columns


# Creating the list for the balance sheet
ls_bs= [] # Create empty list
for l in soup_bs.find_all("div"): 
  #Find all data structure that is ‘div’
  ls_bs.append(l.string) # add each element one by one to the list
    
# Creating the list for the cash flows
ls_cf= [] # Create empty list
for l in soup_cf.find_all("div"): 
  #Find all data structure that is ‘div’
  ls_cf.append(l.string) # add each element one by one to the list
```

```python
# income statement filter
new_ls = list(filter(None,ls))

# Balance sheet filter
new_ls_bs = list(filter(None,ls_bs))

# cash flows filter
new_ls_cf = list(filter(None,ls_cf))


```

```python
# income statemet getting rid of irelavent rows. Make sure to run all above ever time run this
new_ls = new_ls[12:]

# income statemet getting rid of irelavent rows. Make sure to run all above ever time run this
new_ls_bs = new_ls_bs[12:]

# income statemet getting rid of irelavent rows. Make sure to run all above ever time run this
new_ls_cf = new_ls_cf[12:]
```

```python
# zipping the date for income statement
is_data = list(zip(*[iter(new_ls)]*6))

# zipping the date for BS
bs_data = list(zip(*[iter(new_ls_bs)]*5))

# zipping the date for cash Flow
cf_data = list(zip(*[iter(new_ls_cf)]*6))
```

```python
Income_st = pd.DataFrame(is_data[0:])

Balance_sheet = pd.DataFrame(bs_data[0:])


Cash_flows = pd.DataFrame(cf_data[0:])
```

```python
# Cleaning up the Income Statement
Income_st.columns = Income_st.iloc[0] # Name columns to first row of dataframe
Income_st = Income_st.iloc[1:,] # start to read 1st row
# setting the index of each row
Income_st.index=Income_st["Quarterly"].tolist()
Income_st2 = Income_st.drop(["Quarterly", "ttm"], axis=1)
```

```python
# Cleaning the Balance sheet Data
Balance_sheet.columns = Balance_sheet.iloc[0] # Name columns to first row of dataframe
Balance_sheet = Balance_sheet.iloc[1:,] # start to read 1st row
# setting the index of each row
Balance_sheet.index=Balance_sheet["Quarterly"].tolist()
Balance_sheet2 = Balance_sheet.drop(["Quarterly"], axis=1)
```

```python
# Cleaning the Balance sheet Data
Cash_flows.columns = Cash_flows.iloc[0] # Name columns to first row of dataframe
Cash_flows = Cash_flows.iloc[1:,] # start to read 1st row
# setting the index of each row
Cash_flows.index=Cash_flows["Quarterly"].tolist()
Cash_flows= Cash_flows.drop(["Quarterly", "ttm"], axis=1)
```

```python
finalIS = Income_st2
finalBS = Balance_sheet2
finalCF = Cash_flows
```
