```python
# Data pulled from financialModelingPrep API
# import and install dependencies
import requests
import json
import pandas as pd

companyName = "TSLA"

balanceSheetData = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/'+ companyName +'?apikey=dd82553f5e8e6d3a5841a063f3f51223')
incomeStatementData = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/'+ companyName +'?apikey=dd82553f5e8e6d3a5841a063f3f51223')
cashFlowsData = requests.get(f'https://financialmodelingprep.com/api/v3/cash-flow-statement/'+ companyName +'?apikey=dd82553f5e8e6d3a5841a063f3f51223')

balanceSheetObject = balanceSheetData.json()
incomeStatementObject = incomeStatementData.json()
cashFlowsObject = cashFlowsData.json()
```

```python
# converting incomeStatement to DataFrame
incomeStatement=pd.DataFrame(incomeStatementObject).T

# Cleaning up the income Statement Data
incomeStatement=incomeStatement.drop(['period'])
new_header = incomeStatement.iloc[0] 
incomeStatement = incomeStatement[1:] #take the data less the header row
incomeStatement.columns = new_header #set the header row as the df header

# Printing the incomeStatement
incomeStatement
```

```python
# Convert Balance Sheet to DataFrame
balanceSheet=pd.DataFrame(balanceSheetObject).T


# Converting the header
new_header = balanceSheet.iloc[0] 
balanceSheet = balanceSheet[1:] #take the data less the header row
balanceSheet.columns = new_header #set the header row as the df header


balanceSheet


```

```python
# Convert Balance Sheet to DataFrame
cashFlows=pd.DataFrame(cashFlowsObject).T

# Converting the header
new_header = cashFlows.iloc[0] 
cashFlows = cashFlows[1:] #take the data less the header row
cashFlows.columns = new_header #set the header row as the df header

cashFlows

```
